```python
import pandas as pd
```

```python
# 完整的文件路径
file_path = r'D:\EBAC\EBA5002 Business Analytics Practice\EBA2.5 BAP Practice Project\Analysis\Analysis Data.csv'

# 读取CSV文件
df = pd.read_csv(file_path)
```

```python
# 将'order_purchase_timestamp'列转换为日期时间类型
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
# 创建一个新列，提取年份和月份
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month'] = df['order_purchase_timestamp'].dt.month
```

```python
# 使用groupby和pivot_table来计算每年每月每个产品类别的计数数量
result = df.pivot_table(index=['order_year', 'order_month'], columns='product_category_name', values='order_id', aggfunc='count')

# 如果你想要在结果中去掉缺失值，可以使用fillna方法
result = result.fillna(0)

# 重置索引，将年和月作为列
result.reset_index(inplace=True)

# 显示结果
print(result)
```

```python
# 合并 'order_year' 和 'order_month' 为一列，命名为 'year_month'
result['year_month'] = result['order_year'].astype(str) + '-' + result['order_month'].astype(str)

# 删除原有的 'order_year' 和 'order_month' 列
result.drop(columns=['order_year', 'order_month'], inplace=True)

# 重新排列列的顺序，将 'year_month' 放到第一列
result = result[['year_month'] + [col for col in result.columns if col != 'year_month']]
```

```python
# 指定输出文件的完整路径
output_file_path = r'D:\EBAC\EBA5002 Business Analytics Practice\EBA2.5 BAP Practice Project\Analysis\output_result.xlsx'
result.to_excel(output_file_path, index=False)
```
